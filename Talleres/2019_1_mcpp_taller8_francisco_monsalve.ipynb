### 3. [2 puntos]

Usando expresiones regulares, escriba una función de Python que reciba una fecha en formato **Marzo 7, 2019** y retorne la fecha en formato **2019-07-03**

```python
months={"Enero":"01","Febrero":"02","Marzo":"03","Abril":"04","Mayo":"05","Junio":"06","Julio":"07","Agosto":"08","Septiembre":"09","Octubre":"10","Noviembre":"11","Diciembre":"12"}
def date_1(dates):
    year = re.findall("([0-9][0-9][0-9][0-9])", dates)
    año = year[0]
    month = re.findall("^([\w]+).", dates)
    mes = months["".join(month)]
    day = re.findall("([0-9]+),", dates)
    día = "0"+ day[0]
    fecha = año+"-"+día+"-"+mes
    return fecha
```

```python
date_1("Marzo 7, 2019")
```

```python
date_1("Abril 5, 2020")
```

```python
date_1("Diciembre 12, 1996")
```
